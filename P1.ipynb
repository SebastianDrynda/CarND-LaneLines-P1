## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    #return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
            
def draw_line(img, x, y, color=[255, 0, 0], thickness=5):
    """
    Performing a linear fit to x and y using np.polyfit.
    Then, using coefficients of the fit to average the position of each of the lines 
    and extrapolate to the top and bottom of the lane.
    Finally draws it on the image `img` using `color` and `thickness` for the line.
    """
    if len(x) == 0: 
        return
    
    fit = np.polyfit(x, y, 1) 
    max_y = img.shape[0]
    # max_x = img.shape[1]
    
    y1 = max_y
    x1 = int((y1 - fit[1]) / fit[0])
    y2 = int((max_y / 2)) + 50
    x2 = int((y2 - fit[1]) / fit[0])
    cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    #for line in lines:
        #for x1,y1,x2,y2 in line:
            #cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    
    left_x = []
    left_y = []
    right_x = []
    right_y = []

    for line in lines:
        for x1,y1,x2,y2 in line:
            # separating line segments by their slope to decide which segments are part of the left line vs. 
            # the right line
            slope = (y1 - y2) / (x1 - x2)
            if slope < 0:
                left_x.append(x1)
                left_y.append(y1)
                left_x.append(x2)
                left_y.append(y2)
            else:
                right_x.append(x1)
                right_y.append(y1)
                right_x.append(x2)
                right_y.append(y2)

    draw_line(img, left_x, left_y, color, thickness)
    draw_line(img, right_x, right_y, color, thickness) 
            
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    color = [255, 0, 0] 
    thickness = 10
    draw_lines(line_img, lines, color, thickness)
    return line_img

# Python 3 has support for cool math symbols.
def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

## Test Images

Build your pipeline to work on the images in the directory "test_images"

````python
def read_images(image_dir):
    """
    Read all images with the extension *.jpg
    Returns an image list[(filename, image_array)]
    """
    
    image_list = []
 
    for file in glob.glob(image_dir + "/*.jpg"):
        # Read in the image file
        image = mpimg.imread(file)
        image_name = os.path.basename(file)
        # print(file, image_name)
        image_list.append((image_name, image))
        
    return image_list
    
def write_image(image, image_name, image_dir_dest, gray=0, show=0):
    """
    Write an image to destination dir
    If show=1 then the image will be shown
    """
    
    # Create dir if not exist
    if not os.path.exists(image_dir_dest):
        os.makedirs(image_dir_dest)     
        
    filepath_new = image_dir_dest + "/" +  image_name
    if gray:
        plt.imsave(filepath_new, image, cmap='gray')
        if show:
            plt.figure()
            plt.imshow(image, cmap='gray')   
    else:
        plt.imsave(filepath_new, image)
        if show:
            plt.figure()
            plt.imshow(image)   
        
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

## Image Pipeline step 1: Converting images into gray scale 

````python
test_images_list = []
test_images_gray_list = []
test_images_list = read_images("test_images")

def image_pipeline_gray(image):
    return grayscale(image)

for image_name, image in test_images_list:
    image_gray = image_pipeline_gray(image)
    write_image(image_gray, image_name, "test_images_gray", 1, 1)
    test_images_gray_list.append((image_name, image_gray))

print("Image Pipeline step 1: Converting images into gray scale successfully finished")
````

## Image Pipeline step 2: Applying Gaussian smoothing

````python
test_images_gaussian_blur_list = []
blur_kernel_size = 15

def image_pipeline_gaussian_blur(image):
    return gaussian_blur(image, blur_kernel_size)

for image_name, image in test_images_gray_list:
    image_gaussian_blur = image_pipeline_gaussian_blur(image)
    write_image(image_gaussian_blur, image_name, "test_images_gaussian_blur", 1, 1)
    test_images_gaussian_blur_list.append((image_name, image_gaussian_blur))

print("Image Pipeline step 2: Applying Gaussian smoothing successfully finished")
````

## Image Pipeline step 3: Applying Canny transform

````python
test_images_canny_list = []
canny_low_threshold = 20
canny_high_threshold = 100

def image_pipeline_canny(image):
    return canny(image, canny_low_threshold, canny_high_threshold)

for image_name, image in test_images_gaussian_blur_list:
    image_canny = image_pipeline_canny(image)
    write_image(image_canny, image_name, "test_images_canny", 0, 1)
    test_images_canny_list.append((image_name, image_canny))

print("Image Pipeline step 3: Applying Canny transform successfully finished")
````

## Image Pipeline step 4: Applying Region of Interest 

````python
test_images_roi_list = []

def image_pipeline_roi(image):
    y_size = image.shape[0]
    x_size = image.shape[1]    
    vertices = np.array([[0, y_size],[x_size / 2, (y_size / 2) + 20], [x_size, y_size]], dtype=np.int32)
    return region_of_interest(image, [vertices])

for image_name, image in test_images_canny_list:
    image_roi = image_pipeline_roi(image)
    write_image(image_roi, image_name, "test_images_roi", 0, 1)
    test_images_roi_list.append((image_name, image_roi))

print("Image Pipeline step 4: Applying Region of Interest successfully finished")
````

## Image Pipeline step 5: Applying Hough transform

````python
test_images_hough_list = []

rho = 1 # distance resolution in pixels of the Hough grid
theta = np.pi/180 # angular resolution in radians of the Hough grid
threshold = 10     # minimum number of votes (intersections in Hough grid cell)
min_line_length = 30 #minimum number of pixels making up a line
max_line_gap = 1    # maximum gap in pixels between connectable line segments

def image_pipeline_hough(image):
    return hough_lines(image, rho, theta, threshold, min_line_length, max_line_gap)

for image_name, image in test_images_roi_list:
    image_hough = image_pipeline_hough(image)
    write_image(image_hough, image_name, "test_images_hough", 0, 1)
    test_images_hough_list.append((image_name, image_hough))

print("Image Pipeline step 5: Applying Hough transform successfully finished")
````

## Image Pipeline step 6: Merging original image with Hough transform lines

````python
i = 0
for image_name, image in test_images_hough_list:
    image_merged = weighted_img(test_images_list[i][1], image)
    i += 1
    write_image(image_merged, image_name, "test_images_merged", 0, 1)

print("Image Pipeline step 6: Merging original image with Hough transform lines")
````
